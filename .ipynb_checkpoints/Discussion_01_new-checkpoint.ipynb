## II. EDA

````python
# Count missing values by column, excluding full features, and sort by the number of missing entries
missing_all = fifa.isna().sum().reset_index()
missing_all.columns = ["feature", "cnt_na_vals"]
missing_all = missing_all[missing_all["cnt_na_vals"] != 0].sort_values(by="cnt_na_vals", ascending=False).reset_index(drop=True)

# Visualize the 20 least-populated features
twenty_sparsest = missing_all.head(20)
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(data=twenty_sparsest, x="feature", y="cnt_na_vals", ax=ax)
plt.xticks(rotation=90)
plt.xlabel("Features")
plt.ylim(0, 18000)
plt.ylabel("Missing Values")
plt.title("Top 20 Sparsest Columns")

plt.tight_layout()
plt.show()
````

So, from the above, I can see that ```national_team```, ```national_team_position```, ```national_jersey_number```, and ```national_rating``` are all 90% sparse. This makes sense, since  only a handful of elites would be chosen to play for the national team. However, it means that in the case of any Machine Learning, these features would have to be dropped from consideration. Lets subset those populated national entries for later analysis, then move on.

```tags``` and ```traits``` would also likely require consideration, given their sparsity.

````python
# Subset players with National Records.
national = fifa.loc[:, ["id", "national_team", "national_team_position", "national_jersey_number", "national_rating"]].dropna()
national_subset = fifa.loc[fifa["id"].isin(national["id"])]
````

## III. Data Engineering

````python
df = fifa
````

````python
# Sus out columns that are missing less than or equal to 15% of their entries.
fifteen_pct_feats = missing_all.loc[missing_all["cnt_na_vals"] <= 0.15 * fifa.shape[0]].reset_index(drop=True)["feature"].values.tolist()
pos_eng = fifa.loc[:, fifteen_pct_feats]
pos_eng.head(10)
````

````python
pos_eng.info()
````

````python
pos_eng["LM"].astype("str")
````

````python
def calc_from_str(inString):
    """Calculate a sum from a string"""
    val = 0
    if inString.strip().lower() == "nan":
        val = np.nan
    elif "+" in inString:
        a, b = inString.strip().split("+")
        val = int(a) + int(b)
    else:
        val = int(inString)
    return val
````

````python
target_feats_one = pos_eng.loc[:, "LM":"RF"].columns.tolist()
# Display selected columns for transformation
print("Target Feats:", target_feats_one)
# Apply transformation, column-by-column
for col in target_feats_one:
    pos_eng[col] = pos_eng[col].astype("str").apply(calc_from_str)
# Use simple mean imputation to fill missing values
means = pos_eng.loc[:, "LM":"RF"].mean().round(0).reset_index()
means.columns = ["feature", "mean"]
means_dict = dict(zip(means["feature"].values.tolist(), means["mean"].values.tolist()))

pos_eng[target_feats_one] = pos_eng.loc[:, "LM":"RF"].fillna(value=means_dict).astype("int32")
````

````python
pos_eng.info()
````

````python
# Find out which columns still have missing values
still_missing = pos_eng.iloc[:, 26:].columns.tolist()
labels_to_check = still_missing + ["potential", "international_reputation(1-5)"]

# Select those columns
study_set = fifa.loc[:, labels_to_check]

# Use context to determine fill value policies, as appropriate
study_set["club_jersey_number"] = study_set["club_jersey_number"].fillna(value=0)
study_set["club_position"] = study_set["club_position"].fillna(value="Unknown")
study_set["club_team"] = study_set["club_team"].fillna(value="Unknown")

# Review results of latest transformations
study_set
````

````python
imputable_df = study_set.loc[:, ["potential", "international_reputation(1-5)", "release_clause_euro", "value_euro", "wage_euro", "club_rating"]]
imputable_df
````

````python
# Save pre-image of feature statistics for later validation
pre_impute_stats = imputable_df.describe()

# Examine correlation among chosen variables
imputable_features_corr = imputable_df.corr().round(2)
imputable_features_corr
````

It would appear that ```potential``` has a mild to moderate positive correlation with our monetary features ```release_clause_euro```, ```value_euro```, and ```wage_euro```. Likewise, ```international_reputation(1-5)``` is also correlated with these variables, even though the two are not strongly correlated. The goal now, is to utilize known information from the first two columns to impute to "predict" the missing values in the remaining columns. Scikit-learn's experimental ```IterativeImputer``` object utilizes Bayesian Ridge Regression to iteratively impute missing numerical values from all others.

````python
# Impute iteratively
imputer = IterativeImputer(random_state=2025)
imputed_data = imputer.fit_transform(imputable_df)

# Convert back to DataFrame
imputed_df = pd.DataFrame(imputed_data, columns=imputable_df.columns)

imputed_df
````

````python
# Save post-image feature statistics
post_impute_stats = imputed_df.describe()

# Examine the relative change resulting from iterative imputation
rel_change_stats = ((post_impute_stats - pre_impute_stats) / pre_impute_stats).abs().round(2).iloc[:, 2:]
rel_change_stats
````

We can see that, from an aggregate perspective, our data is basically unchanged. That being the case, we can "push" these engineered results back up to ```study_set```.

````python
# Replace columns with missing values with the imputed versions
study_set.loc[:, ["release_clause_euro", "value_euro", "wage_euro", "club_rating"]] = imputed_df.loc[:, ["release_clause_euro", "value_euro", "wage_euro", "club_rating"]]
study_set
````

````python
study_set["contract_end_year"].unique()
````

````python
# Force entry conformity
end_year_dict = {
    '2021': 2021,
    '2020': 2020,
    '2022': 2022,
    '2023': 2023,
    '2024': 2024,
    'Jun 30, 2019': 2019,
    '2025': 2025,
    'nan': np.nan,
    '2019': 2019,
    'May 31, 2020': 2020,
    'Jun 30, 2020': 2020,
    'Dec 31, 2019': 2020,
    'May 31, 2019': 2019,
    'Jan 1, 2020': 2020,
    'May 21, 2019': 2019,
    'Jan 31, 2020': 2020,
    'Feb 27, 2020': 2020,
    'Dec 1, 2019': 2019,
    'Oct 14, 2019': 2019,
    'Dec 31, 2018': 2019,
    'Mar 1, 2019': 2019,
    'Jun 1, 2020': 2020,
    'Jul 2, 2019': 2019,
    'Dec 1, 2020': 2020,
    '2026': 2026
}

study_set["contract_end_year"] = study_set["contract_end_year"].map(end_year_dict)
````

````python
# Visualize contracts
sns.histplot(data=study_set, x="contract_end_year", color="green", shrink=4)
plt.xlabel("Year")
plt.ylabel("Number of Contracts Expiring")
plt.title("Contract Expiry Distribution")

plt.tight_layout()
plt.show()
````

````python
pos_eng.loc[:, study_set.columns.tolist()] = study_set
pos_eng
````

````python
# Coerce integers to 32-bit typing to save memory
for col in ["release_clause_euro", "value_euro", "wage_euro", "club_rating", "potential", "international_reputation(1-5)"]:
    pos_eng[col] = pos_eng[col].astype("int32")

pos_eng.info()
````

````python
# Write our "possible engineering" dataFrame back up to df, which will store clean values.
df.loc[:, pos_eng.columns.tolist()] = pos_eng

temp = df.select_dtypes(include='int64').columns.tolist()
for col in temp:
    if df[col].isna().sum() == 0:
        df[col] = df[col].astype("int32")

df.info()
````

````python
# Subset non-sparse, numerically-interpretable columns from df
X = df.select_dtypes(exclude=['datetime64[ns]', 'object']).drop(labels=["national_jersey_number", "national_rating"], axis=1)
X = X.dropna()
print(f"X Shape: {X.shape}\n")

# Calculate Pearson correlation matrix
corr = X.corr()

# Hierarchical Clustering with Farthest Neighbor method
plt.figure(figsize=(12,5))
dissimilarity = 1 - corr.abs()
Z = linkage(y=squareform(dissimilarity), method='complete')

# Visualize feature hierarchies
dendrogram(Z, labels=X.columns, orientation='top', leaf_rotation=90);
````

The above dendrogram is representative of complete-linkage hierarchical clustering. Each color corresponds to a specific cluster. For example, ```agression```, ```marking```, ```interceptions``` are colored orange and grouped with other offensive features, such as ```stamina```, ```GK_kicking```, etc. Meanwhile, monetary features like ```wage_euro``` and rankings like ```international_ratings(1-5)``` eventually join in with traits like ```composure```. While the code snippet originally used a threshold of ```0.8```, we modified this to reflect the psuedo-minimum cluster height. Within our features, the "orange" cluster reaches a height of ```0.5```, so we choose that as our ```threshold``` below.

````python
# Clusterize the data
threshold = 0.5
labels = fcluster(Z, threshold, criterion='distance')

# Keep the indices to sort labels
labels_order = np.argsort(labels)

# Build a new dataframe with the sorted columns; refactored to avoid looping
clustered = X[X.columns[labels_order]]
corr_new = clustered.corr()

# Generate custom text for hover
hover_text = [
    [f"x: {corr_new.columns[j]};<br>y: {corr_new.index[i]};<br>r: {corr_new.iloc[i, j]:.2f}"
     for j in range(corr_new.shape[1])]
    for i in range(corr_new.shape[0])
]

# Create heatmap
fig = go.Figure(data=go.Heatmap(
    z=corr_new.values,
    x=corr_new.columns,
    y=corr_new.index,
    colorscale='RdBu',
    zmin=-1,
    zmax=1,
    text=hover_text,
    hoverinfo='text'
))

# Layout adjustments
fig.update_layout(
    title="Correlation Matrix",
    width=1000,
    height=1000
)

# Show interactive plot
fig.show()
````

## IV. Insights
Utilizing linkage criteria, we were able to reorder our numerical features, and thus our correlation matrix, so that related features appear closer to one another in terms of their indices. This mimics the functionality of ```method='hclust``` within RLang's ```corrplot``` package. The use of Plotly over Seaborn in this case allows for interactive objects, which also save space, since we don't have to specify ```annot``` within ```sns.heatmap```. Instead, we just hover over cells, and get the details there.

By looking at the above heatmap, we can quickly make a few observations:
* As the "orange" cluster of dendrogram suggested, ```aggression``` is most closely associated with interceptions, tackles, and marking; since $r \ge  0.72$, we can classify these features as having a strong positive correlation with aggression.
* The features ```GK_kicking```, ```GK_reflexes```, and ```GK_handling``` share incredibly strong positive correlations ($ r \ge 0.97$). This unusually high association suggests that these variables may be measuring the same underlying quantity.
* Meanwhile, various positions attacking positions such as Centre Attacking Middle, Left-Winger, and Center Forward show a strong positive correlation with one another. While some positions show $r \approx 0.99$, other remain highly correlated but distinct. Indeed, this indicates yet again that we may have redundant variables.
* Interestingly, ```stamina``` has strong negative correlations with ```GK_handling```, ```GK_reflexes```, and ```GK_kicking```, among others. Players with high stamina tend to be great a dribbling, ball control, and short passes, but may struggle more with precise movements. One could investigate difference between endurance-based support players with high stamina and energetic burst players.
* Some features, such as ```jumping```, ```weak_foot```, and ```contract_year_end``` do not seem to be meaningfully associated with many other variables, if at all.
